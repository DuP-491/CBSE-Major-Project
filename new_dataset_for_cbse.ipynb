```python
import pandas as pd
import os
import random
```

```python
def calculate_adjustment_factor():

    design_document_percentage = random.random()*85+5
    code_percentage = random.random()*80+10
    integration_percentage = random.random()*85+10

    adjustment_factor = 0.4*design_document_percentage + 0.3*code_percentage + 0.3*integration_percentage

    return (design_document_percentage,code_percentage,integration_percentage,adjustment_factor)
```

```python
def calculate_size_equivalent(actual_size,adjustment_factor):
    size_equivalent = (actual_size*adjustment_factor)/100
    return size_equivalent
```

```python
def generate_new_dataset(old_dataset):
    ddl=[]
    cl=[]
    il=[]
    new_size_list=[]
    new_effort_list = []

    new_dataset = pd.DataFrame(columns=list(old_dataset.columns.values))

    print(new_dataset.head())
    for index,row in old_dataset.iterrows():
        for j in range(6):
            
            new_dataset= new_dataset.append(row.to_dict(),ignore_index=True)
            
            dd,c,i,af=calculate_adjustment_factor()
            ddl.append(dd)
            cl.append(c)
            il.append(i)

            size_equivalent = calculate_size_equivalent(row['loc'],af)
            bi = [1.05,1.12,1.20]
            new_effort = int(row['actual']* pow((af/100),bi[j%3]))

            new_effort_list.append(new_effort)
            new_size_list.append(size_equivalent)

    new_dataset=new_dataset.assign(dd=ddl,cod=cl,inte=il,nloc=new_size_list,nactual=new_effort_list)

    return new_dataset
```

```python
old_dataset = pd.read_csv('cocomo81.csv')
new_dataset = generate_new_dataset(old_dataset=old_dataset)
new_dataset.to_csv('dataset_cbse.csv',index=False)
```
